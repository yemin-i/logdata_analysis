## 클래스101의 데이터 분석

### 사전 준비

````python
# 파이썬의 데이터 분석 패키지 판다스(Pandas)를 읽어옵니다.
# 이를 pd라는 축약어로 사용합니다.
import pandas as pd

import warnings
warnings.filterwarnings('ignore')
````

### 데이터 로딩하기

**사용자(```users```) 데이터에 접근하기**

````python
# 판다스(pandas)의 read_csv를 활용하여 클래스101의 사용자(users) 데이터를 저장한 users.csv 를 읽어옵니다.
# 여기서 id 컬럼을 이 데이터의 인덱스(index)로 지정합니다.
# 이 데이터를 users라는 이름의 변수에 할당합니다.
users = pd.read_csv("data/users.csv", index_col = "id")

# users 변수에 할당된 데이터의 행렬 사이즈를 출력합니다.
# 출력은 (row, column) 으로 표시됩니다.
print(users.shape)

# head()로 users 데이터의 상위 5개를 띄웁니다.
users.head()
````

사용자(```users```) 데이터의 각 컬럼에 대한 설명은 다음과 같습니다.

  * ```id``` - 사용자(```users```)의 아이디. 이 컬럼을 인덱스(index)로 활용할 예정입니다.
  * ```email``` - 사용자(```users```)의 이메일 정보입니다. 이메일을 기입하지 않을 가능성이 있으며, 이 경우에는 ```None```이라는 값이 들어갑니다. 개인정보 보호를 위해 가명화(Pseudonymisation) 처리하였습니다.
  * ```phone``` - 사용자(```users```)의 전화번호입니다. 마찬가지로 전화번호를 기입하지 않을 가능성이 있으며, 이 경우에는 ```None```이라는 값이 들어갑니다.  개인정보 보호를 위해 가명화(Pseudonymisation) 처리하였습니다.
  * ```name``` - 사용자(```users```)의 이름입니다. 마찬가지로 개인정보 보호를 위해 가명화(Pseudonymisation) 처리되어 있습니다.
  * ```created_at``` - 사용자(```users```) 데이터를 생성(create)한 날짜입니다.
  * ```updated_at``` - 사용자(```users```) 데이터를 수정(update)한 날짜입니다.
  * ```deleted_at``` - 사용자(```users```) 데이터를 삭제(delete)한 날짜입니다.
  * ```registration_date``` - 사용자(```users```) 회원가입을 한 날짜 정보가 담겨 있습니다. ```created_at``` 컬럼에서 연-월-일만 가져온 데이터입니다.

**클래스(```klasses```) 데이터에 접근하기**

클래스(```klasses```) 데이터의 각 컬럼에 대한 설명은 다음과 같습니다.

  * ```id``` - 클래스(```klasses```)의 아이디. 이 컬럼을 인덱스(index)로 활용할 예정입니다.
  * ```category_id``` - 카테고리(```categories```)의 아이디. 이 컬럼을 활용하여 카테고리(```categories```) 데이터와 연결할 수 있습니다.
  * ```is_deleted``` - 해당 클래스(```klasses```)가 삭제되었으면 1이, 삭제되지 않았으면 0이 기록되어있습니다.
  * ```created_at``` - 클래스(```klasses```) 데이터를 생성(create)한 날짜입니다.
  * ```updated_at``` - 클래스(```klasses```) 데이터를 수정(update)한 날짜입니다.
  * ```deleted_at``` - 클래스(```klasses```) 데이터를 삭제(delete)한 날짜입니다.

**클래스 수강권(```klass_tickets```) 데이터 접근하기**

````python
# 판다스(pandas)의 read_csv를 활용하여 클래스101의 클래스 수강권(klass_tickets) 데이터를 저장한 klass_tickets.csv 를 읽어옵니다.
# 여기서 id 컬럼을 이 데이터의 인덱스(index)로 지정합니다.
# 이 데이터를 klass_tickets라는 이름의 변수에 할당합니다.
klass_tickets = pd.read_csv("data/klass_tickets.csv", index_col = "id")

# klass_tickets 변수에 할당된 데이터의 행렬 사이즈를 출력합니다.
# 출력은 (row, column) 으로 표시됩니다.
print(klass_tickets.shape)

# head()로 klass_tickets 데이터의 상위 5개를 띄웁니다.
klass_tickets.head()
````

클래스 수강권(```klass_tickets```) 데이터의 각 컬럼에 대한 설명은 다음과 같습니다.

  * ```id``` - 클래스 수강권(```klass_tickets```)의 아이디. 이 컬럼을 인덱스(index)로 활용할 예정입니다.
  * ```user_id``` - 사용자(```useres```)의 아이디입니다. 이 컬럼을 활용하여 사용자(```useres```) 데이터와 연결할 수 있습니다.
  * ```klass_id``` - 클래스(```klasses```)의 아이디입니다. 이 컬럼을 활용하여 클래스(```klasses```) 데이터와 연결할 수 있습니다.
  * ```available_days``` - 해당 수강권(```klass_ticketes```)의 유효기간입니다. 일(day)로 기록되어 있습니다. (즉, 300일은 수강권이 개시된 날짜로부터 300일 동안 유효한 수강권이라는 의미)

  * ```created_at``` - 수강권(```klass_ticketes```) 데이터를 생성(create)한 날짜입니다.
  * ```updated_at``` - 수강권(```klass_ticketes```) 데이터를 수정(update)한 날짜입니다.
  * ```finished``` - 수강권(```klass_ticketes```)이 만료되는 날짜입니다.
  * ```deleted_at``` - 수강권(```klass_ticketes```) 데이터를 삭제(delete)한 날짜입니다.

**카테고리(```categories```) 데이터에 접근하기**

````python
# 판다스(pandas)의 read_csv를 활용하여 클래스101의 카테고리(categories) 데이터를 저장한 categories.csv 를 읽어옵니다.
# 여기서 id 컬럼을 이 데이터 의 인덱스(index)로 지정합니다.
# 이 데이터를 categories라는 이름의 변수에 할당합니다.
categories = pd.read_csv("data/categories.csv", index_col = "id")

# categories 변수에 할당된 데이터의 행렬 사이즈를 출력합니다.
# 출력은 (row, column) 으로 표시됩니다.
print(categories.shape)

# head()로 categories 데이터의 상위 5개를 띄웁니다.
categories.head()
````

카테고리(```categories```) 데이터의 각 컬럼에 대한 설명은 다음과 같습니다.

  * ```id``` - 카테고리(```categories```)의 아이디. 이 컬럼을 인덱스(index)로 활용할 예정입니다.

### 데이터 정리

**1. 사용자(```users```) 데이터의 날짜 정보를 ```datetime``` 형태로 변환하겠습니다.**

````python
# pd.to_datetime() 함수를 이용하여 컬럼 형태를 변환하겠습니다.
users['created_at'] = pd.to_datetime(users['created_at'])
users['updated_at'] = pd.to_datetime(users['updated_at'])
users['deleted_at'] = pd.to_datetime(users['deleted_at'])
users['registration_date'] = pd.to_datetime(users['registration_date'])

users.dtypes
````

**2. 클래스(```klasses```) 데이터의 ```is_deleted``` 컬럼 값을 ```0```, ```1```에서 ```True```, ```False```로 바꾸겠습니다.**

````python
# Write your code here!
# replace 함수를 이용해 데이터를 bool 형식으로 변환하겠습니다.
klasses['is_deleted(bool)'] = klasses['is_deleted'].replace(0, False).replace(1, True)

klasses[['is_deleted', 'is_deleted(bool)']].head()
````

**3. 클래스 수강권(```klass_tickets```) 데이터의 ```is_paid``` 컬럼 값을 ```0```, ```1```에서 ```True```, ```False```로 바꾸겠습니다.**

````python
# 데이터를 bool 형식으로 변환하기 위한 함수를 만들겠습니다.
def is_paid_bool(is_paid):
    if is_paid == 1:
        return True
    else:
        return False
````

````python
# 위에서 만든 함수를 적용하겠습니다.
klass_tickets['is_paid(bool)'] = klass_tickets['is_paid'].apply(is_paid_bool)

klass_tickets[['is_paid', 'is_paid(bool)']].head()
````

**4. 클래스(```klasses```) 데이터와 클래스 수강권(```klasses_tickets```) 데이터의 날짜 정보를 datetime 형태로 변환하겠습니다.**

````python
# 먼저, klasses 데이터의 타입을 변경하겠습니다.
klasses['created_at'] = pd.to_datetime(klasses['created_at'])
klasses['updated_at'] = pd.to_datetime(klasses['updated_at'])
klasses['deleted_at'] = pd.to_datetime(klasses['deleted_at'])

klasses.dtypes
````

````python
# 다음으로 klass_tickets 데이터의 타입을 변경하겠습니다.

klass_tickets['created_at'] = pd.to_datetime(klass_tickets['created_at'])
klass_tickets['updated_at'] = pd.to_datetime(klass_tickets['updated_at'])
klass_tickets['finished_at'] = pd.to_datetime(klass_tickets['finished_at'])
klass_tickets['deleted_at'] = pd.to_datetime(klass_tickets['deleted_at'])

klass_tickets.dtypes
````

**5. 수강권(```klass_tickets```) 데이터의 유효기간(```available_days```) 컬럼 값을 정수형(```int```)로 변경하겠습니다.**

````python
# astype()을 이용하여 available_days 컬럼의 데이터 타입을 int형으로 변환하겠습니다.
klass_tickets['available_days(int)'] = klass_tickets['available_days'].astype('int')

klass_tickets[['available_days', 'available_days(int)']].head()
````

### 분석 리포트 생성

**6. 2020년 4월 기준 전체 가입자와 신규 가입자를 구하겠습니다.**

````python
# 총 가입자
total = users.shape[0]

# 2020년 5월 이후 가입자
over_month5 = users.loc[(users['registration_date'].dt.year == 2020) & (users['registration_date'].dt.month > 4)]

# 2020년 4월 기준 총 가입자
total_person = total - over_month5.shape[0]
print(f'전체 가입자 수 : {total_person}명')

# 신규 가입자 수
new_person = users.loc[(users['registration_date'].dt.year == 2020) & (users['registration_date'].dt.month == 4)]
print(f'신규 가입자 수 : {new_person.shape[0]}명')
````

**7.2020년 4월 기준 ```전체 수강생```. 즉, 수강권을 구매한 구매자를 찾겠습니다.**

````python
# 수강권을 구매한 전체 수강생을 구하기 위해 'created_at' 와 'is_paid(bool)' 컬럼을 활용하겠습니다.
paid_klass = klass_tickets.loc[(klass_tickets['created_at'].dt.year == 2020) & (klass_tickets['created_at'].dt.month == 4) 
                               & (klass_tickets['is_paid(bool)'] == True)]

# unique()를 활용하여 여러 개의 수업을 수강한 케이스를 제거하겠습니다.
user_list = list(paid_klass['user_id'].unique())

print(f'2020년 4월 기준 전체 수강생 : {len(user_list)}명')
````

**8. 2020년 4월 30일 기준 유효 수강생. 즉 이미 수업을 듣고 있는 수강생을 구하겠습니다.**

```전체 수강생```과 ```유효 수강생```의 가장 큰 차이점은, "실제 이 서비스(클래스101)을 이용하는 것을 고려하고 있는가?"의 여부입니다. ```전체 수강생```은 수강 신청일(수업을 신청한 날짜)를 기준으로 계산하기 때문에 실제 수업을 듣고 있는 사용자는 반영되지 않을 가능성이 높습니다. 하지만 ```유효 수강생```은 실제 이 서비스를 이용하고 있는 수강생이 몇 명인지를 계산하기 때문에, 서비스의 실사용자가 몇 명인지를 파악할 수 있습니다.

(참고로 교육업이 아닌 다른 서비스에서는, 이와 유사한 지표로 ```DAU(Daily Active Users)``` 또는 ```MAU(Maily Active Users)```를 계산하고 있습니다. ```유효 수강생```은 이와 유사한 개념이라고 보시면 됩니다)

````python
# 'created_at'과 'finished_at' 컬럼을 활용하여 수업을 듣고 있는 수강생 수를 구하겠습니다.
valid_student = klass_tickets[(klass_tickets['created_at'] <= '2020-05-01') & (klass_tickets['finished_at'] > '2020-05-01')]

# unique()를 활용하여 여러 개의 수업을 수강한 케이스를 제거하겠습니다.
valid_student_num = len(valid_student['user_id'].unique())
print(f'2020년 4월 30일 기준 전체 수강생 : {valid_student_num}명')
````

**9. 2020년 4월 이후에 생긴 신규 클래스를 찾고, 이 신규 클래스를 수강 신청한 사용자의 총 인원수를 구하겠습니다.**

````python
# 'created_at' 컬럼을 활용하여 2020년 4월 이후에 생성된 클래스를 수강신청한 사용자 수를 구하겠습니다.
new_class_student = klasses[(klasses['created_at'].dt.year == 2020) & (klasses['created_at'].dt.month >= 4)]

# 위에서 필터링한 user를 klass_tickets의 'klass_id' 컬럼에 매핑하여 총 인원 수를 구하겠습니다.
student_num = klass_tickets[klass_tickets['klass_id'].isin(new_class_student.index)]

print(student_num.shape[0],'명')
````

**10. 2020년 4월 이후에 생긴 신규 클래스 중, ```직무``` 카테고리 클래스를 수강 신청한 사용자의 총 인원수를 구하겠습니다.**

클래스101에는 ```커리어```, ```직무```, ```드로잉```, ```공예```, ```디자인``` 등등 다양한 카테고리의 수업이 있습니다. 이 수업 중 2020년 4월 기준 클래스101이 가장 집중하고 있는 카테고리는 바로 ```직무``` 카테고리입니다.

그런 의미에서, 2020년 4월 이후에 생긴 신규 클래스 중, ```직무``` 카테고리 클래스를 수강 신청한 사용자의 총 인원을 구해보겠습니다.

````python
# categories의 title 컬럼에서 직무로 필터링 하겠습니다. 
job = categories.loc[categories['title'] == '직무']

# klasses의 created_at 컬럼을 2020년 4월 이후로 필터링하겠습니다.
after_four_month_job = klasses[(klasses['created_at'].dt.year == 2020) &  (klasses['created_at'].dt.month >= 4)]

# isin 함수를 활용하여 after_four_month_job의 category_id와 job.index와 매칭되는 케이스의 인덱스를 저장하겠습니다.
after_four_month_job_index = after_four_month_job[after_four_month_job['category_id'].isin(job.index)].index

# isin 함수를 활용하여 klass_tickets의 klass_id와 after_four_month_job_index와 매칭되는 케이스를 저장하겠습니다.
job_students = klass_tickets[klass_tickets['klass_id'].isin(after_four_month_job_index)]

print(job_students.shape[0],'명')
````

### 심화 분석

**11. 월별 신규 가입 사용자(```users```) 수를 구하겠습니다.**

앞서 분석 리포트를 생성할 때는 2020년 4월만을 기준으로 신규 가입 사용자를 구했습니다. 하지만 실제 데이터를 분석할 때는 특정 연월 뿐만 아니라 전체 연월을 계산하여, 이를 통해 신규 가입 사용자가 늘어나고 있는지(=회사가 성장하고 있는지), 내지는 신규 가입 사용자가 늘어나지 않는지(=회사가 정체되어 있는지) 확인합니다. 그런 의미에서 2019년 1월부터 2020년 5월까지의 신규 사용자의 증가 추세를 계산하겠습니다.

````python
# 날짜를 연-월로 변환하는 함수를 만들겠습니다.
def ym(x):
    return str(x)[:7]

users['registration_date(ym)'] = users['registration_date'].apply(ym)
````

````python
# 등록 연-월별 신규 가입자 수를 피벗 테이블로 출력하겠습니다.
month_newbie = pd.pivot_table(data = users, index = 'registration_date(ym)', values = 'name', aggfunc = 'count')
month_newbie
````

**12. 월별 수강권(```klass_tickets```)을 구매한 사용자 수를 구하겠습니다.**

**13. 카테고리(```categories```)별 운영 중인 클래스(```klasses```) 수를 구하겠습니다.**

이번에는 수업 카테고리(```categories```)별 운영중인 클래스(```klasses```)의 개수를 구해보겠습니다. 카테고리별 클래스의 개수를 구할 수 있으면 현재 크리에이터들이 가장 활발하게 수업이 올리는 클래스가 어떤 것인지를 확인할 수 있습니다.

````python
# pd.merge함수를 활용하여 klasses와 categories 데이터를 합치겠습니다. 머지 키는 아래와 같이 설정했습니다.
merged_data = pd.merge(klasses, categories, left_on = 'category_id', right_on = categories.index)

# category_id 컬럼을 id 로 컬럼명을 변환하겠습니다.
merged_data.rename(columns = {'category_id':'id'}, inplace = True)

merged_data.head()
````

````python
# 피벗 테이블을 활용하여 title별 빈도를 출력하겠습니다.
result = pd.pivot_table(merged_data, index = 'title', values = 'id', aggfunc = 'count')

# 고빈도 순으로 정렬하겠습니다.
result.sort_values(by = 'id', ascending = False)
````

**15. 월별, 그리고 클래스 카테고리별 신규 구매자 수를 구하겠습니다.**

마지막으로, 클래스101의 전체 판매 현황을 자세하게 확인할 수 있는 단 하나의 분석 테이블을 계산해보겠습니다. 바로 1) 월별, 2) 카테고리별, 3) 클래스의 신규 구매자 수입니다. 이 지표를 계산할 수 있다면 카테고리별로 클래스 수강생이 늘어나고 있는지(=해당 카테고리가 성장하고 있는지), 정 반대로 클래스 수강생이 늘어나지 않고 정체되고 있는지를 파악 가능합니다.

````python
# klass_tickets와 klasses 데이터를 merge 하겠습니다. merge key는 아래와 같습니다.
total_data = pd.merge(klass_tickets, klasses, left_on = 'klass_id', right_on = klasses.index)

# 신규 구매자 조건이므로 is_paid(bool) 컬럼을  True로 필터링하겠습니다.
total_data = total_data.loc[total_data['is_paid(bool)'] == True].reset_index(drop = True)

total_data.head()
````
